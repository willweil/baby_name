```python
import numpy as np
import pandas as pd
import os.path
from splinter import Browser
```

```python
def print_baby_score(name, avg_score, byj_score, byj_wuxing, byj_jixiong, sz_score, sz_wuxing, sz_jixiong, 
                     mmt_score, mmt_wenhua_score, mmt_wuxing_score, mmt_zhouyi_score, mmt_wuxing, mmt_jieshi, w1518_score):
    print("************************\n      ", name, "\n************************")
    print("卜易居：", byj_score, "五行：", byj_wuxing, "吉凶: ", byj_jixiong)
    print("三藏算命：", sz_score, "五行：", sz_wuxing, "吉凶: ", sz_jixiong)
    print("1518：", w1518_score)
    print("美名腾:", mmt_score, "文化印象", mmt_wenhua_score, "五行八字", mmt_wuxing_score, 
          "周易", mmt_zhouyi_score, "五行：", mmt_wuxing)
    print("名字解释：", mmt_jieshi)
    print("平均分：{0:.1f}".format(avg_score))
    return None
```

```python
def baby_score_aggregator(name_lst, file_to_save=r"C:\Users\WEIL\Documents\AI_ML\baby_name\baby_name_scores.xlsx"):
    if os.path.isfile(file_to_save):
        saved_results = pd.read_excel(file_to_save, header=0, sheet_name='baby_name_scores')
        name_lst_new = []
        for name in name_lst:
            if name in saved_results.name.values:
                avg_score = saved_results.query("name == @name").avg_score.values[0]
                byj_score = saved_results.query("name == @name").byj_score.values[0]
                byj_wuxing = saved_results.query("name == @name").byj_wuxing.values[0]
                byj_jixiong = saved_results.query("name == @name").byj_jixiong.values[0]
                sz_score = saved_results.query("name == @name").sz_score.values[0]
                sz_wuxing = saved_results.query("name == @name").sz_wuxing.values[0]
                sz_jixiong = saved_results.query("name == @name").sz_jixiong.values[0]
                mmt_score = saved_results.query("name == @name").mmt_score.values[0]
                mmt_wenhua_score = saved_results.query("name == @name").mmt_wenhua_score.values[0]
                mmt_wuxing_score = saved_results.query("name == @name").mmt_wuxing_score.values[0]
                mmt_zhouyi_score = saved_results.query("name == @name").mmt_zhouyi_score.values[0]
                mmt_wuxing = saved_results.query("name == @name").mmt_wuxing.values[0]
                mmt_jieshi = saved_results.query("name == @name").mmt_jieshi.values[0]
                w1518_score = saved_results.query("name == @name").w1518_score.values[0]                
                print_baby_score(name, avg_score, byj_score, byj_wuxing, byj_jixiong, sz_score, sz_wuxing, sz_jixiong,
                                 mmt_score, mmt_wenhua_score, mmt_wuxing_score, mmt_zhouyi_score, mmt_wuxing, mmt_jieshi, w1518_score)
            else:
                name_lst_new.append(name)
        if not name_lst_new:
            return None
    name_lst = name_lst_new
    container = []
    for name in name_lst:
        with Browser('firefox', headless=True, executable_path=r'C:\Program Files\geckodriver\geckodriver.exe') as browser:    
            #############################
            # 卜易居
            #############################
            website = 'http://xmcs.buyiju.com/dafen.php'
            # browser.driver.set_window_size(640, 480)
            browser.visit(website)
            last_name_xpath = '/html/body/div[2]/div[1]/div[2]/div[3]/form/input[1]'
            first_name_xpath = '/html/body/div[2]/div[1]/div[2]/div[3]/form/input[2]'
            run_button = '/html/body/div[2]/div[1]/div[2]/div[3]/form/input[3]'
            input_last_name = browser.find_by_xpath(last_name_xpath)[0]
            input_first_name = browser.find_by_xpath(first_name_xpath)[0]
            run_button = browser.find_by_xpath(run_button)[0]
            input_last_name.fill("赖")
            input_first_name.fill(name)
            run_button.click()
            browser.windows.current = browser.windows.current.next
            score_xpath = '/html/body/div[2]/div[1]/div[2]/div[4]/p[35]/b/font'
            byj_score = float(browser.find_by_xpath(score_xpath).text)
            wuxing1_xpath = '/html/body/div[2]/div[1]/div[2]/div[4]/table/tbody/tr/td[1]/table/tbody/tr[5]/td[4]'
            jixiong1_xpath = '/html/body/div[2]/div[1]/div[2]/div[4]/table/tbody/tr/td[1]/table/tbody/tr[6]/td[4]'
            wuxing1 = browser.find_by_xpath(wuxing1_xpath).text
            jixiong1 = browser.find_by_xpath(jixiong1_xpath).text
            wuxing2 = ''
            jixiong2 = ''
            if len(name) == 2:
                wuxing2_xpath = '/html/body/div[2]/div[1]/div[2]/div[4]/table/tbody/tr/td[1]/table/tbody/tr[5]/td[5]'
                jixiong2_xpath = '/html/body/div[2]/div[1]/div[2]/div[4]/table/tbody/tr/td[1]/table/tbody/tr[6]/td[5]'
                wuxing2 = browser.find_by_xpath(wuxing2_xpath).text
                jixiong2 = browser.find_by_xpath(jixiong2_xpath).text
            byj_wuxing = wuxing1 + wuxing2
            byj_jixiong = jixiong1 + jixiong2

        with Browser('firefox', headless=True, executable_path=r'C:\Program Files\geckodriver\geckodriver.exe') as browser:    
            #############################
            # 三藏算命
            #############################
            website = 'http://www.sheup.com/xingming_dafen.php'
            browser.visit(website)
            last_name_xpath = '/html/body/div/div[2]/div[2]/div[2]/div[3]/div[2]/form/p[2]/input[1]'
            first_name_xpath = '/html/body/div/div[2]/div[2]/div[2]/div[3]/div[2]/form/p[2]/input[2]'
            run_button_xpath = '/html/body/div/div[2]/div[2]/div[2]/div[3]/div[2]/form/p[2]/input[3]'
            input_last_name = browser.find_by_xpath(last_name_xpath).first
            input_first_name = browser.find_by_xpath(first_name_xpath).first
            run_button = browser.find_by_xpath(run_button_xpath).first
            input_last_name.fill("赖")
            input_first_name.fill(name)
            run_button.click()
            score_xpath = '/html/body/div/div[2]/div[2]/div[2]/div[5]/div[2]/div[1]/span'
            sz_score = float(browser.find_by_xpath(score_xpath).text)
            wuxing1_xpath = '/html/body/div/div[2]/div[2]/div[2]/div[7]/div[2]/table/tbody/tr[3]/td[4]'
            jixiong1_xpath = '/html/body/div/div[2]/div[2]/div[2]/div[7]/div[2]/table/tbody/tr[3]/td[6]/font'
            wuxing1 = browser.find_by_xpath(wuxing1_xpath).text
            jixiong1 = browser.find_by_xpath(jixiong1_xpath).text
            wuxing2 = ''
            jixiong2 = ''
            if len(name) == 2:
                wuxing2_xpath = '/html/body/div/div[2]/div[2]/div[2]/div[7]/div[2]/table/tbody/tr[4]/td[4]'
                jixiong2_xpath = '/html/body/div/div[2]/div[2]/div[2]/div[7]/div[2]/table/tbody/tr[4]/td[6]/font'
                wuxing2 = browser.find_by_xpath(wuxing2_xpath).text
                jixiong2 = browser.find_by_xpath(jixiong2_xpath).text
            sz_wuxing = wuxing1 + wuxing2
            sz_jixiong = jixiong1 + jixiong2

        with Browser('firefox', headless=True, executable_path=r'C:\Program Files\geckodriver\geckodriver.exe') as browser:    
            #############################
            # 1518
            #############################
            website = 'http://www.1518.com/'
            browser.visit(website)
            name_xpath = '//*[@id="main"]/div[1]/div[1]/div[1]/table/tbody/tr/td[3]/input'
            run_button = '//*[@id="main"]/div[1]/div[1]/div[1]/table/tbody/tr/td[8]/input'
            input_name = browser.find_by_xpath(name_xpath)[0]
            run_button = browser.find_by_xpath(run_button)[0]
            input_name.fill("赖"+name)
            run_button.click()
            score_xpath = '/html/body/div[3]/div[3]'
            w1518_score = float(browser.find_by_xpath(score_xpath).text)

        with Browser('firefox', headless=True, executable_path=r'C:\Program Files\geckodriver\geckodriver.exe') as browser:  
            #############################
            # 美名腾
            #############################
            website = 'https://www.meimingteng.com/yy/'
            browser.visit(website)
            last_name_xpath = '//*[@id="ctl00_ContentPlaceHolder1_InputBasicInfo1_tbXing"]'
            first_name_xpath = '//*[@id="ctl00_ContentPlaceHolder1_InputBasicInfo1_tbMingWords"]'
            gender_xpath = '//*[@id="ctl00_ContentPlaceHolder1_InputBasicInfo1_ddlGenders"]/option[2]'
            year_xpath = '//*[@id="ctl00_ContentPlaceHolder1_InputBasicInfo1_ddlYear"]/option[118]'
            month_xpath = '//*[@id="ctl00_ContentPlaceHolder1_InputBasicInfo1_ddlMonth"]/option[9]'
            day_xpath = '//*[@id="ctl00_ContentPlaceHolder1_InputBasicInfo1_ddlDay"]/option[21]'
            hour_xpath = '//*[@id="ctl00_ContentPlaceHolder1_InputBasicInfo1_ddlHour"]/option[24]'
            minute_xpath = '//*[@id="ctl00_ContentPlaceHolder1_InputBasicInfo1_ddlMinute"]/option[27]'
            input_last_name = browser.find_by_xpath(last_name_xpath)[0]
            input_first_name = browser.find_by_xpath(first_name_xpath)[0]
            input_gender = browser.find_by_xpath(gender_xpath)[0]
            input_year = browser.find_by_xpath(year_xpath)[0]
            input_month = browser.find_by_xpath(month_xpath)[0]
            input_day = browser.find_by_xpath(day_xpath)[0]
            input_hour = browser.find_by_xpath(hour_xpath)[0]
            input_minute = browser.find_by_xpath(minute_xpath)[0]    
            input_last_name.fill("赖")
            input_first_name.fill(name)
            input_gender.click()
            input_year.click()
            input_month.click()
            input_day.click()
            input_hour.click()
            input_minute.click()
            try:
                browser.find_by_xpath('//*[@id="btSubmit"]').first.click()   # click run button
            except:
                try: 
                    browser.find_by_xpath('//*[@id="btSubmit"]')[0].click()   # click run button
                except:
                    raise
            wenhua_score_xpath = '//*[@id="ctl00_ContentPlaceHolder1_ShowNameDetails1_lbNameScore"]/font[2]'
            mmt_wenhua_score = float(browser.find_by_xpath(wenhua_score_xpath).first.text)
            wuxing_score_xpath = '//*[@id="bdAppSummDiv"]/table[5]/tbody/tr/td/font[1]/b'
            mmt_wuxing_score = float(browser.find_by_xpath(wuxing_score_xpath).first.text)
            zhouyi_score_xpath = '//*[@id="zhouyi"]/span[2]'
            mmt_zhouyi_score = browser.find_by_xpath(zhouyi_score_xpath).first.text
            if not mmt_zhouyi_score:
                mmt_zhouyi_score = 0
                mmt_score = (mmt_wenhua_score + mmt_wuxing_score)/2
            else:
                mmt_zhouyi_score = float(mmt_zhouyi_score[:-1])
                mmt_score = (mmt_wenhua_score + mmt_wuxing_score + mmt_zhouyi_score)/3
            wuxing1_xpath = '//*[@id="ctl00_ContentPlaceHolder1_ShowNameDetails1_lbName"]/table/tbody/tr[3]/td[6]'
            wuxing1 = browser.find_by_xpath(wuxing1_xpath).first.text
            wuxing2 = ''
            if len(name) == 2:
                wuxing2_xpath = '//*[@id="ctl00_ContentPlaceHolder1_ShowNameDetails1_lbName"]/table/tbody/tr[4]/td[6]'
                wuxing2 = browser.find_by_xpath(wuxing2_xpath).first.text
            mmt_wuxing = wuxing1 + wuxing2
            mmt_jieshi = browser.find_by_xpath('//*[@id="ctl00_ContentPlaceHolder1_ShowNameDetails1_lbNameMeaning"]').text
            mmt_jieshi = mmt_jieshi.replace(' ', '').replace('\n', '')


        avg_score = float((byj_score + sz_score + mmt_score) / 3.0)        
        print_baby_score(name, avg_score, byj_score, byj_wuxing, byj_jixiong, sz_score, sz_wuxing, sz_jixiong, 
                         mmt_score, mmt_wenhua_score, mmt_wuxing_score, mmt_zhouyi_score, mmt_wuxing, mmt_jieshi, w1518_score)

        row = [name, avg_score, byj_score, byj_wuxing, byj_jixiong, sz_score, sz_wuxing, sz_jixiong, 
               mmt_score, mmt_wenhua_score, mmt_wuxing_score, mmt_zhouyi_score, mmt_wuxing, mmt_jieshi, w1518_score]
        container.append(row)
        col_name = ['name', 'avg_score', 'byj_score', 'byj_wuxing', 'byj_jixiong', 'sz_score', 'sz_wuxing', 'sz_jixiong', 
                    'mmt_score', 'mmt_wenhua_score', 'mmt_wuxing_score', 'mmt_zhouyi_score', 'mmt_wuxing', 'mmt_jieshi', 'w1518_score']
        results_df = pd.DataFrame(container, columns=col_name).astype('str')
        if not os.path.isfile(file_to_save):
            results_df.to_excel(file_to_save, encoding='utf_8_sig', header=True, index=False)
        # update the file if there is any new entry
        if os.path.isfile(file_to_save):
            saved_results = pd.read_excel(file_to_save, header=0, sheet_name='baby_name_scores')
            saved_results = pd.concat([saved_results, results_df], ignore_index=True)
            saved_results.drop_duplicates(subset=['name'],inplace=True)
            saved_results.to_excel(file_to_save, encoding='utf_8_sig', header=True, index=False)
    return (results_df, saved_results)
```
